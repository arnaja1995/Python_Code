```python
""" Demo for 'Newton's Method' as an assignment for numerical analysis by Arnaja Mitra """

import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x)
def f(x):
    return np.sin(x**2) - x**3 - 1
# Define the derivative f'(x) of f(x)
def fp(x):
    return 2*x*np.cos(x**2) - 3*x**2 
    
# Define left and right interval boundaries for plotting
left_interval  = -2
right_interval =  2

# Define x grid of points for computation and plotting the function
xpart = np.linspace(left_interval, right_interval, num=100)
# Define y values for the given function
ypart = f(xpart)

# Form a list of starting points for Newton's method to loop over, which is coded later 
# Consider all x values as starting points
pointlist = xpart

# Figure plotting
plt.figure()
# plot the function f(x)
plt.plot(xpart,ypart)
# Plot the zero line to understand better
plt.hlines(0, left_interval, right_interval)
# Labelling the axes
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$f(x)$", fontsize=16)
# Set limits for the x and y axes 
# For x limits
plt.xlim((left_interval, right_interval))
# For y limits for convergence
plt.ylim((-10,50))

# Loop over all starting points defined in pointlist
for x0 in pointlist:
    # Set desired precision and max number of iterations
    prec_goal = 1.e-10
    nmax = 1000
    # Initialize values for iteration loop
    reldiff = 1
    xi = x0
    counter = 0
    # Start iteration loop
    while reldiff > prec_goal and counter < nmax:
        
        # Get the number of necessary iterations at x0
        # Compute relative difference
        reldiff = np.abs(f(xi)/fp(xi)/xi)
        # Compute next xi
        x1 = xi - f(xi)/fp(xi)
        # Considering the output as an input for next iteration step
        xi = x1
        # Increase the counter
        counter += 1
    # Plot the number of necessary iterations at that particular x0
    plt.plot(x0,counter,'r.', markersize=5)
    # Print the output
    print(x0,counter,reldiff)
    
```
